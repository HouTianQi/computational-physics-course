```python
import math 
from matplotlib import pyplot as plt
import numpy as np




def mfunction(L):
    m=0
    mlist=[1.0/math.sqrt(i**2+j**2+k**2)*(-1)**(i+k+j) for i in range(-L,L+1) for j in range(-L,L+1) for k in range(-L,L+1) if not (i==k==j==0)]
    for x in mlist:
        m=m+x
        
    return m

x=range(5,80)
y=[abs(mfunction(xi)) for xi in x]

plt.loglog(x,y,color='red')
plt.show()
```
