```python
import numpy as np
from matplotlib import pyplot as plt
import scipy
import math

x=np.linspace(0,10,100)
j0=scipy.special.jn(0,x)
j1=scipy.special.jn(1,x)


plt.figure(figsize=(4,8))
plt.plot(x,j0,label=' order=0',color='red')
plt.plot(x,j1,label='order=1',color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def jnsum(n,x,N):
    result=0.0
    for m in range(N):
        result=result+(-1)**m/math.factorial(m)/math.factorial(m+n)*(x/2.0)**(2*m+n)
        
    return result

x=np.linspace(0,10,100)
j0sum=jnsum(0,x,50)
j1sum=jnsum(1,x,50)

plt.figure(figsize=(8,4))
plt.plot(x,j0sum,label='jo',color='red')
plt.plot(x,j1sum,label='j1',color='blue')
plt.legend()
plt.show()
```

```python
def jnint(n,x,tau):
    return np.cos(n*tau-x*np.sin(tau))/np.pi


def tra(n,x,N):
    a=0.0
    b=np.pi
    h=(b-a)/N
    s=(jnint(n,x,a)+jnint(n,x,b))/2.0
    for k in range(1,N):
        s=s+jnint(n,x,a+k*h)
    s=s*h
    
    return s
x=np.linspace(0,10,100)
jntra0=tra(0,x,100)
jntra1=tra(1,x,100)
plt.figure(figsize=(4,8))
plt.plot(x,jntra0,label='tra0',color='red')
plt.plot(x,jntra1,label='tra1',color='blue')
plt.legend()
plt.show()
    
```

```python
def simpson(n,x,N):
    a=0.0
    b=np.pi
    h=(b-a)/N
    s=jnint(n,x,a)+jnint(n,x,b)
    for i in range(1,N//2):
        s=s+4*jnint(n,x,a+(2*i-1)*h)
    for k in range(1,N//2-1):
        s=s+2*jnint(n,x,a+2*k*h)
        
    return s 

x=np.linspace(0,5,100)
jnsim0=simpson(0,x,100)
jnsim1=simpson(1,x,100)

plt.figure(figsize=(4,8))
plt.plot(x,jnsim0,label='simpson 0',color='red')
plt.plot(x,jnsim1,label='simpson 1',color='blue')
plt.legend()
plt.show()

```
